# 3D Hallucination Discriminator Demo

This notebook demonstrates how to use the discriminator functions to analyse 3D models for possible geometric hallucinations. You can adjust the file paths to point to your own OBJ, STL or PLY files.

```
import numpy as np
import pandas as pd
import trimesh
import discriminator  # assumes discriminator.py is in the same directory

# optional: reload discriminator if you edit it
import importlib
importlib.reload(discriminator)
```

```
# Specify the paths to your 3D models here.
# Replace these with your own files if desired.
model_paths = [
    '/opt/pyvenv/lib/python3.13/site-packages/gradio/media_assets/models3d/Bunny.obj',
    '/opt/pyvenv/lib/python3.13/site-packages/gradio/media_assets/models3d/face.obj',
    '/opt/pyvenv/lib/python3.13/site-packages/gradio/media_assets/models3d/sofia.stl',
]
```

```
results = []
for path in model_paths:
    topo, spec, score = discriminator.analyse_model(path)
    results.append({
        'model': path.split('/')[-1],
        'components': int(topo['components']),
        'euler_chi': topo['euler_char'],
        'genus': topo['genus'],
        'eigen_variance': spec['eigen_variance'],
        'fiedler_value': spec['fiedler_value'],
        'hallucination_score': score,
    })

# Convert to DataFrame for display
import pandas as pd
results_df = pd.DataFrame(results)
results_df
```

## Interpretation

- **components**: number of connected components. A value greater than 1 indicates disconnected pieces and may suggest hallucination.
- **euler_chi** and **genus**: topological invariants. Genus > 0 means the surface has holes or handles; these may be unwanted in certain models.
- **eigen_variance** and **fiedler_value**: spectral metrics from the Laplacian. High variance and very low Fiedler value suggest irregular or loosely connected geometry.
- **hallucination_score**: aggregated heuristic score between 0 (clean) and 1 (severe hallucination).

Feel free to adjust thresholds in `discriminator.py` to calibrate the score for your dataset.
